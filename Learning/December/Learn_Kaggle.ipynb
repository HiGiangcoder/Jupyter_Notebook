```python
import pandas as pd
import requests
import json
```

### Đọc file trên link github cá nhân và lưu vào dataMelb

```python
url = "https://raw.githubusercontent.com/HiGiangcoder/AI/d1fd6ef4e3912edccd184c1453cdaaa7ee3a320f/melb_data.csv"
dataMelb = pd.read_csv(url)

# Mô tả dữ liệu đầu vào
dataMelb.describe()
```

### Loại bỏ các giá trị bằng 0 (.dropma)

```python
dataMelb = dataMelb.dropna(axis=0)
```

### Lấy Prediction Target

```python
y = dataMelb.Price
print(y)
```

### Get features

```python
Melb_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = dataMelb[Melb_features]
X.describe()
# X.head()
```

### Vẽ Scatter plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Chuyển thành dataFrame để dễ thao tác
df = pd.DataFrame(X, columns=Melb_features)
df['Price'] = y
# Vẽ pairplot

sns.pairplot(df)
plt.show()
```

## Decision Tree model

### Building your model
1. **Define:** Xác định loại Model sẽ sử dụng(chẳng hạn như Decision tree).
2. **Fit:** Lấy patterns được cung cấp.
3. **Predict:** Dự đoán kết quả bài toán.
4. **Evaluate:**: Xác định độ chính xác của sự dựa đoán bằng model trên như thế nào.


### Chia data thành 2 tập:
1. X_train, y_train
2. X_test, y_test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=69)

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

### Tạo và chạy model
- Decision tree từ thư viện sklearn.tree và train model đó bởi `model.fit(X_train, y_train)`

```python
from sklearn.tree import DecisionTreeRegressor

melb_model = DecisionTreeRegressor(random_state=69)
melb_model.fit(X_train, y_train)
```

### Dự đoán các test sau khi train xong model

```python
price_predict = melb_model.predict(X_test)
```

### Tính toán độ lỗi của model bằng MSE và MAE

```python
from sklearn.metrics import mean_squared_error as MSE, mean_absolute_error as MAE

print(f'MSE = {MSE(y_test, price_predict)}')
print(f'MAE = {MAE(y_test, price_predict)}')
```

### Đánh giá Sự Underfitting và Overfitting

```python
import numpy as np

def get_mae(max_leaf_nodes, X_train, X_test, y_train, y_test):
  model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=96)
  model.fit(X_train, y_train)
  preds = model.predict(X_test)

  return MAE(y_test, preds)

arr_leaf = np.array(range(2, 1000))
arr_preds = []
arr_test = []

for max_leaf_nodes in arr_leaf:
  my_mae = get_mae(max_leaf_nodes, X_train, X_test, y_train, y_test)
  #print(f"Max leaf nodes: {max_leaf_nodes} \t\t Mean Absolute Error: {my_mae}")
  arr_preds.append(my_mae)
  arr_test.append(get_mae(max_leaf_nodes, X_train, X_train, y_train, y_train))

arr_preds = np.array(arr_preds)
arr_test = np.array(arr_test)
```

### Vẽ đồ thị mô tả sự biến thiên của max_leaves_nodes đối với độ lỗi MAE của train và test

```python
plt.plot(arr_leaf, arr_preds, label='predict test data', color='blue')
plt.plot(arr_leaf, arr_test, label='predict train data', color='red')

plt.legend(loc='upper right')
plt.xlabel('Max leaf nodes')
plt.ylabel('MAE')
plt.show()
```

## Random Forests model
- Thuộc nhóm ensemble learning.
- Random forests gồm nhiều cây decision tree để tính toán.
- 1. Nếu đáp án thuộc dạng regression, đáp án cuối cùng là trung bình cộng của nhiều cây đó.
- 2. nếu đáp án thuộc dạng Classification(phân loại), đáp án cuối cùng là nhãn có đa số phiếu bầu từ các cây.

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(max_leaf_nodes=1000, random_state=696)
forest_model.fit(X_train, y_train)
price_preds = forest_model.predict(X_test)

print(f'MAE = {MAE(y_test, price_preds)}')
```
